Calibration
===

* Proper calibration was not done before
    * Previously, I just zoomed into each of the peaks and estimated the centroid and used those points to find the slope and offset.
* Now, I found the peaks using `TSpectrum`, and fit gaussians to those peaks and extracted the positions of the centers of the peaks
    * I then used these values to find the slope and offset of the Silicon detectors.
* The final result is that the energies shifted slightly. Not enough to redo the cuts (from what I can observe).
